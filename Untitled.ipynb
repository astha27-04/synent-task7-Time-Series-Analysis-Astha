```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load dataset
df = pd.read_csv("stocks.csv")
df.head(2)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#cleaning the dataset
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')
df = df.dropna()
```

```python
df.shape
```

```python
# EXTRA COLUMNS - FOR POWER BI!
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['month_name'] = df['date'].dt.strftime('%B')
df['day_of_week'] = df['date'].dt.dayofweek
df['day_name'] = df['date'].dt.strftime('%A')
df['quarter'] = df['date'].dt.quarter
```

```python
df.to_csv('stocks_cleaned.csv', index=False)

print("✅ Done! File saved as 'stocks_cleaned.csv'")
print(f"Uur data has {len(df)} rows and {len(df.columns)} columns")
```

```python
# Calculate 30-day average
df['trend'] = df['last_value'].rolling(30).mean()
```

```python
# Chart 1: Trend
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
df['trend'] = df['last_value'].rolling(30).mean()
plt.plot(df['date'], df['last_value'])
plt.plot(df['date'], df['trend'], 'r', linewidth=2)
plt.title('Price Trend')
```

```python
# Chart 2: Seasonality
plt.subplot(1, 2, 2)
df['month'] = df['date'].dt.month
monthly = df.groupby('month')['last_value'].mean()
plt.bar(monthly.index, monthly.values)
plt.title('Monthly Average')

plt.tight_layout()
plt.show()
```
